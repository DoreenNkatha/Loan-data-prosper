```python
###Data cleaning
##Selecting columns meaningful to this analysis
cols = ['LoanStatus','BorrowerRate','BorrowerAPR','Occupation','EmploymentStatus','IncomeRange','StatedMonthlyIncome','LoanOriginalAmount','MonthlyLoanPayment','ProsperRating (Alpha)','LoanOriginationDate','ProsperScore','CreditGrade']
Loan = loans[cols]
Loan.head()
```

```python
Loan.info()
Loan['CreditGrade'].unique()
Loan['EmploymentStatus'].unique()
Loan['IncomeRange'].unique()
Loan['Occupation'].unique()
Loan['ProsperRating (Alpha)'].unique()

```

```python
## convert loan originationdate to datetime 
Loan['LoanOriginationDate']= pd.to_datetime(Loan['LoanOriginationDate'])
Loan.info()
```

```python
Loan['ProsperRating (Alpha)'] = Loan['ProsperRating (Alpha)'].fillna('NaN')
Loan['ProsperRating (Alpha)'].unique()
```

```python
ordinal_var_dict ={'ProsperRating (Alpha)' : ['NaN', 'A', 'D', 'B', 'E', 'C', 'AA', 'HR'], 'IncomeRange': ['$25,000-49,999', '$50,000-74,999', 'Not displayed', '$100,000+',
       '$75,000-99,999', '$1-24,999', 'Not employed', '$0']}
for var in ordinal_var_dict:
              ordered_var = pd.api.types.CategoricalDtype(ordered= True, categories=ordinal_var_dict[var])
              Loan[var] = Loan[var].astype(ordered_var)
              
```

```python
Loan['ProsperRating (Alpha)'].value_counts()
```

```python
Loan['IncomeRange'].value_counts()
```

```python
Loan.isnull().sum()
```

```python
#Replacing nun with not available in occupation.
Loan.Occupation = Loan.Occupation.fillna("Not available")
Loan['Occupation'].isnull().sum()
Loan.info()
```

```python
# Univariate Plots
## Occupation
plt.figure(figsize=[7,10])
base_color = sb.color_palette()[0]
sb.countplot(data = Loan, x = 'LoanStatus', color=base_color )
plt.xticks(rotation = 90)
plt.show()
```

```python
# Income Range
plt.figure(figsize=[7,10])
base_color = sb.color_palette()[0]
sb.countplot(data = Loan, x = 'IncomeRange', color=base_color )
plt.xticks(rotation = 45)
plt.title("Income Range")
plt.xlabel('Income Range')
plt.ylabel('Count')
plt.show()
```

```python
#Occupation 
plt.figure(figsize=[7,10])
base_color = sb.color_palette()[0]
sb.countplot(data = Loan, y = 'Occupation', color=base_color )
plt.title('Occupation')
plt.xlabel('count')
plt.ylabel('occupation')

plt.show()
```

```python
# loan original amount
plt.figure(figsize = [10,8])
binsize =1000
bins = np.arange(Loan['LoanOriginalAmount'].min(), Loan['LoanOriginalAmount'].max()+binsize, binsize)
plt.hist(data = Loan, x= 'LoanOriginalAmount', bins= bins)
plt.title('Loan Original Amount')
plt.xlabel('Loan original amount')
plt.ylabel('Count')
```

The original amount is skewed to the right, to better understand the data we convert the histogram to log scale 


```python
logbin = 0.025
bins = 10 ** np.arange(3, np.log10(Loan['LoanOriginalAmount'].max()) + logbin, logbin)
plt.figure(figsize = [10,8])
plt.xscale('log')
plt.xticks([500, 1e3, 2e3,3e3, 5e3, 1e4, 2e4, 3e4, 5e4], ['500', '1k', '2k', '3k', '5k', '10k', '20k', '30k', '50K'])
plt.hist(data = Loan, x= 'LoanOriginalAmount', bins=bins)
plt.title('Loan Original amount (log')
plt.xlabel('loan orginal amount in dollarrs')
plt.ylabel('count')
plt.show()
```

The highest amount is between 3k and 5k. 

```python
# Stated monthly income
plt.figure(figsize = [10,5])
binsize= 50
bins= np.arange (0, Loan['MonthlyLoanPayment'].max() + binsize, binsize)
plt.hist(data = Loan, x= 'MonthlyLoanPayment', bins= bins)
plt.title('MonthlyLoanPayment')
plt.xlabel('MonthlyLoanPayment')
plt.ylabel('Count')
```

The monthly repayment is right skewed, for better understanding we convert it the scale to log

```python
logbin=0.025
bins = 10** np.arange(1, np.log10(Loan['MonthlyLoanPayment'].max())+logbin, logbin)
plt.figure(figsize=[10,8])
plt.hist(data = Loan, x= 'MonthlyLoanPayment', bins= bins)
plt.xscale('log')
plt.xticks([10, 20, 50, 100, 200, 500, 1e3, 2e3], ['10', '20','50', '100', '200', '500', '1000', '2000'])
plt.title('Monthly payment log scale')
plt.xlabel(' Monthly loan payment($)')
plt.ylabel('count')
```

Monthly loan payment is normally distributed with the highest amount being between 100 and 200.
